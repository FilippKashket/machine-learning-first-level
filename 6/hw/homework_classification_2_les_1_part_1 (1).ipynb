### 1.10
Для каждого классификатора выведите матрицу ошибок и самостоятельно рассчитайте метрику recall (расчеты должны присутствовать). Проверьте расчеты, воспользовавшись встроенной метрикой recall_score.

В нашем случае выживание - это 1, не выживание - это 0. Незнаю как определить метрику потому что не совсем понятен смысл исследования. В целом тут важны обе метрики и какую-то одну более важную выбрать сложно. Скорее всего надо выбирать Полноту т.к. надо минимизировать число ложно утонувших. Если мы не слишком уверены, то лучше посчитать человека выжившим чем что он утонул.

```python
from sklearn.metrics import confusion_matrix, recall_score
```

```python
def mtrx_error(y_test, y_predicted):
    tp = ((y_test == y_predicted)&(y_predicted==1)).sum()
    tn = ((y_test == y_predicted)&(y_predicted==0)).sum()
    fp = ((y_test != y_predicted)&(y_predicted==1)).sum()
    fn = ((y_test != y_predicted)&(y_predicted==0)).sum()

    return {'tp':tp, 'tn':tn, 'fp':fp, 'fn':fn}
```

```python
nn_predicted = knn_model.predict(x_test)

nb_predicted = nb_model.predict(x_test)

dtc_predicted = dtc_model.predict(x_test)

lg_predicted = lg_model.predict(x_test)

print('knn')
nn_errors = mtrx_error(y_test, nn_predicted) 
print(nn_errors)
nn_recall = nn_errors['tp']/(nn_errors['tp']+nn_errors['fn']) 
print(f'recall {nn_recall}')
nn_recall_lib = recall_score(y_test, nn_predicted)
print(f'recall lib {nn_recall_lib}')

print('Naive Bayes')
nb_errors = mtrx_error(y_test, nb_predicted)
print(nb_errors)
nb_recall = nb_errors['tp']/(nb_errors['tp']+nb_errors['fn']) 
print(f'recall {nb_recall}')
nb_recall_lib = recall_score(y_test, nb_predicted)
print(f'recall lib {nb_recall_lib}')

print('Tree')
dtc_errors = mtrx_error(y_test, dtc_predicted)
print(dtc_errors)
dtc_recall = dtc_errors['tp']/(dtc_errors['tp']+dtc_errors['fn']) 
print(f'recall {dtc_recall}')
dtc_recall_lib = recall_score(y_test, dtc_predicted)
print(f'recall lib {dtc_recall_lib}')

print('Log reg')
lg_errors = mtrx_error(y_test, lg_predicted)
print(lg_errors)
lg_recall = lg_errors['tp']/(lg_errors['tp']+lg_errors['fn']) 
print(f'recall {lg_recall}')
lg_recall_lib = recall_score(y_test, lg_predicted)
print(f'recall lib {lg_recall_lib}')
```
