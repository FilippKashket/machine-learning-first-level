&nbsp;&nbsp;&nbsp;&nbsp;
<br><br>
&nbsp;&nbsp;&nbsp;&nbsp; Любой бизнес хочет максимизировать количество клиентов. Для достижения этой цели важно не только пытаться привлечь новых, но и удерживать уже существующих. Удержать клиента обойдется компании дешевле, чем привлечь нового. Кроме того, новый клиент может оказаться слабо заинтересованным в услугах бизнеса и с ним будет сложно работать, тогда как о старых клиентах уже есть необходимые данные по взаимодействию с сервисом. 
 <br>

&nbsp;&nbsp;&nbsp;&nbsp; Соответственно, прогнозируя отток, мы можем вовремя среагировать и попытаться удержать клиента, который хочет уйти. Опираясь на данные об услугах, которыми пользуется клиент, мы можем сделать ему специальное предложение, пытаясь изменить его решение об уходе от оператора. Благодаря этому задача удержания будет легче в реализации, чем задача привлечения новых пользователей, о которых мы еще ничего не знаем.<br>

&nbsp;&nbsp;&nbsp;&nbsp; Вам предоставлен набор данных от телекоммуникационной компании. В данных содержится информация о почти шести тысячах пользователей, их демографических характеристиках, услугах, которыми они пользуются, длительности пользования услугами оператора, методе оплаты, размере оплаты. 
<br>

&nbsp;&nbsp;&nbsp;&nbsp; Cтоит задача проанализировать данные и спрогнозировать отток пользователей (выявить людей, которые продлят контракт и которые не продлят). Работа должна включать в себя следующие обязательные пункты: 
1. Описание данных *(с расчетом базовых статистик)*,

3. Построение моделей для прогнозирования оттока *(с обоснованием выбора той или иной модели)* на основе проверенных гипотез и выявленных взаимосвязей, 
4. Сравнение качества полученных моделей. 

Во втором разделе обязательно должно присутствовать обоснование гипотез, подробное описание выявленных взаимосвязей, а также их визуализация.

# 1. Описание данных

```python
import pandas as pd

#Загрузим данные
data = pd.read_csv('telecom_users.csv')
data.head()
```

```python
data.sort_values(by=['TotalCharges']).head()
```

```python
#Преобразуем данные
data['gender'] = data['gender'].map({'Male':1, 'Female':0})
data['Partner'] = data['Partner'].map({'Yes':1, 'No':0})
data['Dependents'] = data['Dependents'].map({'Yes':1, 'No':0})
data['PhoneService'] = data['PhoneService'].map({'Yes':1, 'No':0})
data['MultipleLines'] = data['MultipleLines'].map({'Yes':1, 'No':0,'No phone service':2})
data['InternetService'] = data['InternetService'].map({'No':0,'DSL':1,'Fiber optic':2})
data['OnlineSecurity'] = data['OnlineSecurity'].map({'Yes':1, 'No':0,'No internet service':2})
data['OnlineBackup'] = data['OnlineBackup'].map({'Yes':1, 'No':0,'No internet service':2})
data['DeviceProtection'] = data['DeviceProtection'].map({'Yes':1, 'No':0,'No internet service':2})
data['TechSupport'] = data['TechSupport'].map({'Yes':1, 'No':0,'No internet service':2})
data['StreamingTV'] = data['StreamingTV'].map({'Yes':1, 'No':0,'No internet service':2})
data['StreamingMovies'] = data['StreamingMovies'].map({'Yes':1, 'No':0,'No internet service':2})
data['Contract'] = data['Contract'].map({'Month-to-month':0, 'One year':1, 'Two year':2})
data['PaperlessBilling'] = data['PaperlessBilling'].map({'Yes':1, 'No':0})
data['PaymentMethod'] = data['PaymentMethod'].map({'Electronic check':0, 'Mailed check':1, 'Bank transfer (automatic)':2, 'Credit card (automatic)':3})
data['Churn'] = data['Churn'].map({'Yes':1, 'No':0})
data['TotalCharges'] = data['TotalCharges'].apply(lambda x: "0" if x==" " else x).astype('float')
```

```python
#Сохраним айдишники, чтобы потмо понять какой юзер 
#IDs = data['customerID']

#Для начала удалим Unnamed: 0 
data = data.drop(columns=["Unnamed: 0"])

```

```python
#Сначала посмотрим есть ли пропущенные значения
data.info()
```

Посчитаем:
- количество мужчин и женщин
- количество отоков по отношению к общему числу клиентов
- средний, минимальный, максимальный и медианный месячный платеж
- дополнительную аналитику(количество подключенных услуг), а затем среднее, минимальное, максимальное и медианное количество услуг
- средние, минимальное, максимальное и медианное время пользования сервисом

```python
#количество мужчин и женщин
[women, men] = data.groupby(['gender'])['gender'].count()

print(f'Женщин {women}, Мужчин {men} \n')
#Относительно пола, данные практически симметричные

#количество оттоков по отношению к общему числу клиентов
churn = data['Churn'].sum()/data['customerID'].count()*100
print(f'За весь период отток составил {churn:0.4} % \n')

#средний, минимальный, максимальный и медианный месячный платеж
print('Месячный платеж')
print(data['MonthlyCharges'].agg(['min','max','mean','median']) )
print('\n')

#дополнительную аналитику(количество подключенных услуг), а затем среднее, минимальное, 
#максимальное и медианное количество услуг
data['services'] = (data['PhoneService']==1).astype('int')+(data['MultipleLines']==1).astype('int')+\
                (data['OnlineSecurity']==1).astype('int')+(data['OnlineBackup']==1).astype('int')+\
                (data['DeviceProtection']==1).astype('int')+(data['TechSupport']==1).astype('int')+\
                (data['StreamingTV']==1).astype('int')+(data['StreamingMovies']==1).astype('int')
print('Количество услуг')
print(data['services'].agg(['min','max','mean','median']) )  
print('\n')

#средние, минимальное, максимальное и медианное время пользования сервисом
print('Время пользования сервиса')
print(data['tenure'].agg(['min','max','mean','median']) )
print('\n')
```

# 2. Исследование зависимостей и формулирование гипотез

```python
import matplotlib.pyplot as plt
import seaborn as sns

#Уберем айдишники, для графиков они не нужны
data_for_plot = data.drop(columns=["customerID"])

colormap = plt.cm.RdBu
plt.figure(figsize=(14,12))
plt.title('Взаимосвязи параметров', y=1.05, size=15)

sns.heatmap(data_for_plot.astype(float).corr(),linewidths=0.2,vmax=2.0, 
            square=True, cmap=colormap, linecolor='white', annot=True)
```

```python
#Видно, что наличие различных подключенных сервисов значительно коррелирует друг с другом. Я это предположил заранее 
#и специлаьно сделал фичу services с количеством подключенных сервисов, поэтому фичи с информацией по отдельным сервисам 
#мы удалим


data_for_plot_2 = data_for_plot.drop(columns=['MultipleLines','PhoneService','OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport',
                                             'StreamingTV', 'StreamingMovies'])
```

```python
g = sns.pairplot(data_for_plot_2, 
                 hue='Churn', palette = 'seismic', size=1.2, plot_kws=dict(s=10))
g.set(xticklabels=[])
```

Не совсем понятно какие гипотезы тут надо формулировать. Фичи наши влияют на то уйдет человек или нет. Очевидно, что прогнозирвания нужно использовать модели классификации. Возьмем те которые мы использовали в последней домашке про Kaggle:
- GradientBoostingClassifier
- RandomForestClassifier

# 3. Построение моделей для прогнозирования оттока

```python
import numpy as np
from sklearn.model_selection import (GridSearchCV,
                                     train_test_split,
                                     StratifiedKFold)

from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression

#Подготовим дата сеты для обучения
train_targets = data_for_plot_2['Churn']

train_values = data_for_plot_2.drop(columns=['Churn']) 

x_train, x_test, y_train, y_test = train_test_split(train_values, 
                                                    train_targets,
                                                    test_size=0.3,
                                                    random_state=17)


# параметры базовых алгоритмов
[learning_rate] = {'learning_rate': np.arange(0.1, 0.6, 0.1)} # GradientBoostingClassifier

rfc_params = {'n_estimators': range(10, 100, 10), # RandomForestClassifier
              'min_samples_leaf': range(1, 5)}

svc_params = {'kernel': ['linear', 'rbf'], # SVC
              'C': np.arange(0.1, 1, 0.2)}

lr_params = {'C': np.arange(0.5, 1, 0.1)}

skf = StratifiedKFold(n_splits=5, random_state=17)


#Обучим модели
gbc = GradientBoostingClassifier(random_state=17)
grid_gbc = GridSearchCV(gbc, param_grid=gbc_params, cv=skf)
model_gbc = grid_gbc.fit(x_train, y_train)
gbc_learning_rate = model_gbc.best_params_['learning_rate']

rfc = RandomForestClassifier(random_state=17)
grid_rfc = GridSearchCV(rfc, param_grid=rfc_params, cv=skf)
rfc_model = grid_rfc.fit(x_train, y_train)
rfc_n_estimators = rfc_model.best_params_['n_estimators']
rfc_min_samples_leaf = rfc_model.best_params_['min_samples_leaf']

svc = SVC(random_state=17)
grid_svc = GridSearchCV(svc, param_grid=svc_params, cv=skf)
svc_model = grid_svc.fit(x_train, y_train)
svc_kernel = svc_model.best_params_['kernel']
svc_C = svc_model.best_params_['C'] 

lr = LogisticRegression(random_state=17)
grid_lr = GridSearchCV(lr, param_grid=lr_params, cv=skf)
lr_model = grid_lr.fit(x_train, y_train)
lr_C = lr_model.best_params_['C']  

#Считаем предсказания на проверочном дата сете
gbc_base = GradientBoostingClassifier(random_state=17, learning_rate=gbc_learning_rate)
model_gbc_base = gbc_base.fit(x_train, y_train)
gbc_base_predict = model_gbc_base.predict(x_test)

rfc_base = RandomForestClassifier(random_state=17, n_estimators=rfc_n_estimators, min_samples_leaf=rfc_min_samples_leaf)
model_rfc_base = rfc_base.fit(x_train, y_train)
rfc_base_predict = model_rfc_base.predict(x_test)

svc_base = SVC(random_state=17, kernel=svc_kernel, C=svc_C)
model_svc_base = svc_base.fit(x_train, y_train)
svc_base_predict = svc_base.predict(x_test)

lr_base = LogisticRegression(random_state=17, C=lr_C)
model_lr_base = lr_base.fit(x_train, y_train)
lr_base_predict = model_lr_base.predict(x_test)
```

# 4. Сравнение качества моделей 

```python
from sklearn.metrics import accuracy_score
gbc_accuracy =accuracy_score(gbc_base_predict, y_test)
print(f'GradientBoostingClassifier accuracy: {gbc_accuracy}')

rfc_accuracy = accuracy_score(rfc_base_predict, y_test)
print(f'RandomForestClassifier accuracy: {rfc_accuracy}')

svc_accuracy = accuracy_score(svc_base_predict, y_test)
print(f'SVC accuracy: {svc_accuracy}')

lr_accuracy = accuracy_score(lr_base_predict, y_test)
print(f'LogisticRegression accuracy: {lr_accuracy}')
```
